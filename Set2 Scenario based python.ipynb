```python
#Write logic to compute the sum of the digits of a given number.
pc=int(input("Enter the pass code(4-digits): "))
sum =0
while pc>0:
    digit= pc%10 # it will give the reminder
    sum=sum+digit # 0+reminder
    pc=pc//10 # integer division of 10(omit the fractional part)
print("The sum of the digit is: ",sum)


```

```python
#Write logic to take a number and return its reverse.
num=int(input("Enter the number: "))
reversed_number = 0
while(num>0):
    last_digit = num%10
    reversed_number = reversed_number*10+last_digit
    num = num//10
print("The reversed number is: ",reversed_number)
```

```python
#logic to check whether a given number is an Armstrong number.
num=int(input("Enter the number: "))
n = num
power = len(str(num))
total = 0

while n > 0 :
    digit = n % 10
    total += digit ** power
    n //= 10
    
if total == num:
    print("its a Armstrong number")
else:
    print("not a Armstrong number")    
```

```python
#logic to perform this operation on a given string(swapping the 1st & last character of the string)
s = input("Enter a string: ")

# Check length to avoid errors
if len(s) > 1:
    new_str = s[-1] + s[1:-1] + s[0]
else:
    new_str = s  # single character remains same

print("Swapped string:", new_str)
```

```python
#Write logic to convert a given decimal number into its binary equivalent.
n = int(input("Enter the number: "))
binary = ""
if n == 0:
    binary = "0"    
while n > 0:
    r = n % 2
    b = str(r) + binary
    n = n// 2    
    print("Binary numbers:",b)   
```

```python
#Write logic to find the longest word in a sentence.
s = input("Enter a sentence: ")
words = s.split()

longest_word = words[0]

for w in words:
    if len(w) > len(longest_word):
        longest_word = w

print("Longest word:", longest_word)
```

```python
#Write logic to check whether two given strings are anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

# Preprocess strings
s1 = s1.replace(" ", "").lower()
s2 = s2.replace(" ", "").lower()

if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not an Anagram")
```
